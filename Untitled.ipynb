```python
# KHAI BÁO CÁC THƯ VIỆN CẦN SỬ DỤNG
import os
import cv2
import matplotlib.pyplot as plt
from skimage import color,data
from skimage.filters import threshold_local
from skimage.filters import threshold_otsu
from PIL import Image as im
import numpy as np
from skimage.filters import try_all_threshold
```

$$
Khai \ bao \ ham \ hien \ thi \ anh
$$

```python
def show_image(image, title='Image', cmap_type='gray'):
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

$$
Cau \ 1: \ Hien \ thi \ kich \ thuoc \ anh
$$

```python
path1=r'C:\Users\NguyenHien\TTXLA\BTAP_CHUONG1\marid.jpg'
image=plt.imread(path1)
shape=image1.shape
size=image.size
print('Shape hinh :',shape)
print('Size hinh :',size)
show_image(image,'Image')
```

$$
Bai \ 2: \ Chuyen \ doi \ anh \ mau \ sang \ anh \ xam
$$

```python
# Chuyển đổi ảnh thành ảnh xám
gray_scaled = color.rgb2gray(image)
# Hiển thị ảnh gốc
show_image(image, 'Original RGB image')
# Hiển thị ảnh xám
show_image(gray_scaled, 'Grayscale image')
```

$$
Cau \ 3: Lat \ nguoc \ anh
$$

```python
#Lật ảnh theo chiều dọc
seville_vertical_flip = np.flipud(image)
#Lật ảnh theo chiều ngang
horizontally_flipped = np.fliplr(image)
#Hiển thị ảnh đã lật
show_image(seville_vertical_flip,'Ảnh lật theo chiều dọc')
show_image(horizontally_flipped,'Ảnh lật theo chiều ngang')
```

Cau \ 4: Luoc \ do \ muc \ xam

```python
red_channel = image[:, :, 0]
# Vẽ lược đồ kênh màu đỏ với khoảng màu [0, 256]
plt.hist(red_channel.ravel(), bins=256)
# Đặt tên cho khung hình và hiển thị
plt.title('Red Histogram')
plt.figure()
plt.axis('off')  
plt.imshow(red_channel)
plt.figure()
```

$$
Cau \ 5: Phan \ nguong \ toan \ cuc
$$

```python
path2=r'C:\Users\NguyenHien\TTXLA\BTAP_CHUONG1\anh.jpg'
image2=plt.imread(path2)
# Chuyển ảnh màu sang ảnh xám bằng hàm rgb2gray
anhxam = color.rgb2gray(image2)
# Tính toán ngưỡng bằng phương pháp otsu
thresh = threshold_otsu(anhxam)
# Lấy ngưỡng cho ảnh xám
binary = anhxam >= thresh
# Hiển thị ảnh kết quả nhị phân
show_image(anhxam, ' Gray scale image')
show_image(binary, 'Binary image')
```

$$
Cau \ 6: Phan \ nguong \ cuc \ bo
$$


```python
local_thresh = threshold_local(anhxam, block_size=35, offset=10)
# Apply local thresholding and obtain the binary image  
binary_local = anhxam > local_thresh
show_image(binary_local, 'binary_local image')
```

```python
print([anhxam.min(),anhxam.max()])
```

```python
#Phân ngưỡng ảnh với offset=0.1
local_thresh1 = threshold_local(anhxam, block_size=35, offset=0.1)
# Apply local thresholding and obtain the binary image  
binary_local1 = anhxam > local_thresh1
show_image(binary_local1, 'binary_local image')
```

```python
#Phân ngưỡng ảnh với offset=0.05
local_thresh2 = threshold_local(anhxam, block_size=35, offset=0.05)
# Apply local thresholding and obtain the binary image  
binary_local2 = anhxam > local_thresh2
show_image(binary_local2, 'binary_local image')
```
